## 1. What to import, how to conduct basic plot

```python
import numpy as np
import matplotlib.pyplot as plt

pi = np.pi
x = np.linspace(-4 * pi, 4 * pi, 1000)
y = np.sin(x)/x
plt.plot(x, y)
plt.show()
```

## 2. How to set up Figure

```python
x = np.linspace(-1, 1, 50)
y1 = 2 * x
y2 = 3 * x

plt.figure(num = 1, figsize = (8,5)) #设置figure1
plt.plot(x, y1)

plt.figure(num = 2, figsize = (5,5)) #设置figure2
plt.plot(x, y2)
plt.plot(x, y1, color = "red", linewidth = 3.0 , linestyle = '--')  #plot时可以设置figure的颜色，线宽和风格等等

plt.show()
```

## How to set up X-axis

```python
x = np.linspace( -1, 1, 1000)
y = 2 * x

plt.figure(num = 1)
plt.plot(x, y)

plt.xlim((-4, 4)) #设置x轴显示区间
plt.ylim((-4, 4)) #设置y轴显示区间

plt.xlabel('This is X axis') #设置y轴标签
plt.ylabel('This is Y axis') #设置x轴标签


plt.figure(num = 2)
plt.plot(x, y)

plt.xticks(np.linspace(-4, 4, 5))                                   #设置横轴记号，-4到4共5个点
plt.yticks([-4, -2, 0, 2, 4],
           ['really bad', 'bad','normal', 'good', 'really good'])   #设置竖轴记号，共5个点对应不同文字

plt.figure(num = 3)
plt.plot(x, y)

plt.xticks(np.linspace(-4, 4, 5))                                                    #设置横轴记号，-4到4共5个点
plt.yticks([-4, -2, 0, 2, 4],
           [r'$really \ bad$', r'$bad \ \beta$',r'$normal$', r'$good$', r'$really \ good$'])  #用 LaTex 把横轴文字字体改好看一点
                                                                                      # 格式为  r'$ 内容 $'
                                                                                      # 空格和数学符号比如Pi,beta，需在前面加转制符 \


plt.show()
```

## Move Spines

_坐标轴线和轴上的记号组合在一起被称为脊柱（Spines，一条线段上有一系列的凸起，是不是很像脊柱骨啊~），它记录了数据区域的范围。它们可以放在任意位置，不过上方的例子，我们都把它放在图的四边。_

_实际上每幅图有四条脊柱（上下左右），为了将脊柱放在图的中间，我们必须将其中的两条（上和右）设置为无色，然后调整剩下的两条到合适的位置——数据空间的 0 点。_

```python
x = np.linspace(-5, 5, 500)
y = 3 * x
plt.plot(x, y)

ax = plt.gca()   #gca = 'get current axis'

ax.spines['right'].set_color('none')   #右边轴颜色改为无
ax.spines['top'].set_color('none')     #上边轴颜色改为无

ax.xaxis.set_ticks_position('bottom')  #设下边轴为X轴
ax.yaxis.set_ticks_position('left')    #设左边轴为y轴

ax.spines['bottom'].set_position(('data', 0))  #把下边轴移到Y轴0刻度
ax.spines['left'].set_position(('data', 0))    #把左边轴移到x轴0刻度

plt.show()
```

**set_position(position) 函数定义**

**set the position of the spine**

Spine position is specified by a 2 tuple of (position type, amount). The position types are:

- 'outward' : place the spine out from the data area by the specified number of points. (Negative values specify placing the spine inward.)
- 'axes' : place the spine at the specified Axes coordinate (from 0.0-1.0).
- 'data' : place the spine at the specified data coordinate.

Additionally, shorthand notations define a special positions:

- 'center' -> ('axes',0.5)
- 'zero' -> ('data', 0.0)
