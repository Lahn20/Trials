## 1. What to import, how to conduct basic plot

```python
import numpy as np
import matplotlib.pyplot as plt

pi = np.pi
x = np.linspace(-4 * pi, 4 * pi, 1000)
y = np.sin(x)/x
plt.plot(x, y)
plt.show()
```

```python
x = np.linspace( -1, 1, 1000)
y = 2 * x
plt.plot(x, y)
plt.show()
```

## 2. How to set up Figure

```python
x = np.linspace(-1, 1, 50)
y1 = 2 * x
y2 = 3 * x

plt.figure(num = 1, figsize = (8,5)) #设置figure1
plt.plot(x, y1)

plt.figure(num = 2, figsize = (5,5)) #设置figure2
plt.plot(x, y2)
plt.plot(x, y1, color = "red", linewidth = 3.0 , linestyle = '--')  #可以设置figure的颜色，线宽和风格

plt.show()
```
