## 1. What to import, how to conduct basic plot

```python
import numpy as np
import matplotlib.pyplot as plt

pi = np.pi
x = np.linspace(-4 * pi, 4 * pi, 1000)
y = np.sin(x)/x
plt.plot(x, y)
plt.show()
```

## 2. How to set up Figure

```python
x = np.linspace(-1, 1, 50)
y1 = 2 * x
y2 = 3 * x

plt.figure(num = 1, figsize = (8,5)) #figure1
plt.plot(x, y1)

plt.figure(num = 2, figsize = (5,5)) #figure2
plt.plot(x, y2)
plt.plot(x, y1, color = "red", linewidth = 3.0 , linestyle = '--')  #plotfigure

plt.show()
```

## How to set up X-axis

```python
x = np.linspace( -1, 1, 1000)
y = 2 * x

plt.figure(num = 1)
plt.plot(x, y)

plt.xlim((-4, 4)) #x
plt.ylim((-4, 4)) #y

plt.xlabel('This is X axis') #y
plt.ylabel('This is Y axis') #x


plt.figure(num = 2)
plt.plot(x, y)

plt.xticks(np.linspace(-4, 4, 5))                                   #-445
plt.yticks([-4, -2, 0, 2, 4],
           ['really bad', 'bad','normal', 'good', 'really good'])   #5

plt.figure(num = 3)
plt.plot(x, y)

plt.xticks(np.linspace(-4, 4, 5))                                                    #-445
plt.yticks([-4, -2, 0, 2, 4],
           [r'$really \ bad$', r'$bad \ \beta$',r'$normal$', r'$good$', r'$really \ good$'])  # LaTex 
                                                                                      #   r'$  $'
                                                                                      # Pi,beta \


plt.show()
```

## Move Spines

_Spines~_

_ 0 _

```python
x = np.linspace(-5, 5, 500)
y = 3 * x
plt.plot(x, y)

ax = plt.gca()   #gca = 'get current axis'

ax.spines['right'].set_color('none')   #
ax.spines['top'].set_color('none')     #

ax.xaxis.set_ticks_position('bottom')  #X
ax.yaxis.set_ticks_position('left')    #y

ax.spines['bottom'].set_position(('data', 0))  #Y0
ax.spines['left'].set_position(('data', 0))    #x0

plt.show()
```

**set_position(position) **

**set the position of the spine**

Spine position is specified by a 2 tuple of (position type, amount). The position types are:

- 'outward' : place the spine out from the data area by the specified number of points. (Negative values specify placing the spine inward.)
- 'axes' : place the spine at the specified Axes coordinate (from 0.0-1.0).
- 'data' : place the spine at the specified data coordinate.

Additionally, shorthand notations define a special positions:

- 'center' -> ('axes',0.5)
- 'zero' -> ('data', 0.0)
